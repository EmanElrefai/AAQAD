# **AAQAD Meger Tool**

---

**This is  AAQAD Merger Tool**

**- It has an objective to merge all AAQAD parts**

**- AAQAD parts are named: 'AAQAD-vVersionNumber_PartNumber'**

**- AAQAD parts begin with part 0 as default**

**- You need to specify number of AAQAD part in AAQAD_PARTS_NUMBER varibale**

---

```
## Upload all AAQAD parts from Google Drive into 
#  'data' folder to Colab 

from google.colab import drive

drive.mount('/content/drive')
# drive source where all parts are uploaded
src = '/content/drive/My Drive/MRQA_Arabic Research - Adel Atef/Data Set Collection/AAQAD V1.0/'
```

```
## Varibale to be set with total number of required parts to merge
#  AAQAD parts at part 0 by default

AAQAD_PARTS_NUMBER = 23
```

```
import json

def AAQAD_Meger():
  
  ## important declarations
  # final dic collecting all parts together
  AAQAD_dic = {}
  AAQAD_dic["version"] = "v1.0"
  AAQAD_dic['data'] = []
  #used to assign correct IDs
  Question_ID = 1
  article_count = -1
  
  for part_number in range(0,AAQAD_PARTS_NUMBER):
    print("\n-> Merging AAQAD Part: ",part_number," <-")
  
    ## load current JSON part
    with open(src+'AAQAD-v1.0 ('+str(part_number)+').json') as json_file:  
      file = json.load(json_file)
      
      ## looping on articles ...
      for item in file["data"]:
        ############## Start of Each Article Loop ##############
        print("####################################################")
        article_count += 1
        print("Article no ",article_count, " : ",item["title"])

        ## looping on paragraphs
        parag_count = -1
        article_paragraphs = []
        for parag in item["paragraphs"]:
          ############## Start of Each Paragraph Loop ##############
          parag_count += 1
          print("\nPragraph no: ",parag_count)

          paragraph = parag["context"]
          print("-------------------------------")

          ## looping on questions
          ques_count = -1
          qas = []
          for ques in parag["qas"]:
            ############## Start of Each Question Loop ##############
            ques_count += 1
            print("\nquestion no: ",ques_count)

            question = ques["question"]          

            is_impossible = ques["is_impossible"]

            if is_impossible == False:
              ans_text = ques["answers"][0]["text"]
              ans_start= ques["answers"][0]["answer_start"]
            else:
              if len(ques["plausible_answers"]) > 0:
                ans_text = ques["plausible_answers"][0]["text"]
                ans_start= ques["plausible_answers"][0]["answer_start"]
              else:
                #handle empty plausible answer section
                ans_text = ""
                ans_start= -1

            # saving in qas array
            ans_details = {'text': ans_text,
                           'answer_start':ans_start }          
            if is_impossible == False:
              qas.append({'question' : question,
                          'id': Question_ID,
                          'answers': [ans_details],
                          'is_impossible': False})
            else:
              if len(ques["plausible_answers"]) > 0:
                qas.append({'plausible_answers':[ans_details],
                          'question' :  ques["question"],
                          'id': Question_ID,
                          'answers': [],
                          'is_impossible': True})
              else:
                #handle empty plausible answer section
                qas.append({'plausible_answers':[],
                          'question' :  ques["question"],
                          'id': Question_ID,
                          'answers': [],
                          'is_impossible': True})  

            Question_ID += 1
            ############## End of Each Question Loop ##############  

          article_paragraphs.append({'qas': qas, 'context' :paragraph })      
          ############## End of Each Paragraph Loop ##############  

        AAQAD_dic['data'].append({'title' : item["title"],'paragraphs': []})
        AAQAD_dic['data'][article_count]['paragraphs'] = article_paragraphs

        print("####################################################")
        ############## End of Each Article Loop ##############
  
   ############## End of Each JSON Part Loop ##############
    
    
  ## save final dic into a json file & download it
  print("Saving & Download Final AAQAD Version !")
  with open('./AAQAD-v1.0.json', 'w') as outfile:
    json.dump(AAQAD_dic, outfile)
  
  
  return
```

# **Testing [ Run Here ]**

```
AAQAD_Meger()
```
